```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python

def conversion(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
conversion('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(conversion)
```

```python
movies['keywords']= movies['keywords'].apply(conversion)
```

```python
movies.head()
```

```python
def conversion_crew(obj):
  L=[]
  count=0
  for i in ast.literal_eval(obj):
    if count != 3:
      L.append(i['name'])
      count+=1
    else: break
  return L
```

```python
movies['cast']=movies['cast'].apply(conversion_crew)
```

```python
movies.head()
```

```python
def fetch(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job'] =='Director':
      L.append(i['name'])
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
final_df=movies[['movie_id','title','tags']]
```

```python
final_df.head()
```

```python
final_df['tags']=final_df['tags'].apply(lambda x:" ".join(x))
```

```python
final_df.head()
```

```python
final_df['tags']=final_df['tags'].apply(lambda x:x.lower())
```

```python
final_df.head()
```

# Leave

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)  
```

```python
final_df['tags']=final_df['tags'].apply(stem)
```

```python
final_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(final_df['tags']).toarray()
```

```python
vectors
```

```python
#cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommend(movie):
  movie_index=final_df[final_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(final_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
